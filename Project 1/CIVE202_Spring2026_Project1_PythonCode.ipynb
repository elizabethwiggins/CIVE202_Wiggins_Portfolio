```python
import pandas as pd

air_data = pd.read_csv('CIVE202_Spring2026_Project1_AirData.csv')

air_data = pd.DataFrame(air_data)

```

```python
air_data_grouped = air_data.groupby(['sensor.name'])

mean_air_data_aggregated_pm_2_5 = air_data_grouped.agg(mean_pm_2_5 = ('pm2.5_atm', 'mean'))
                                               
                                                 
top5_mean_pm_2_5 = (
    mean_air_data_aggregated_pm_2_5
    .sort_values(by = 'mean_pm_2_5', ascending = False)
    .head(5))



top5_mean_pm_2_5
```

```python
median_air_data_aggregated_pm_2_5 = air_data_grouped.agg(median_pm_2_5 = ('pm2.5_atm', 'median'))
                                               
                                                 
top5_median_pm_2_5 = (
    median_air_data_aggregated_pm_2_5
    .sort_values(by = 'median_pm_2_5', ascending = False)
    .head(5))



top5_median_pm_2_5
```

```python
def assign_humidity(humidity):
    if humidity < 50 :
        return 'Low'
    elif 50 <= humidity <= 80:
        return 'High'
    elif humidity > 81: 
        return 'Very High'
    else:
        return '-'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

air_data  
```

```python
def assign_temperature(temperature):
    if temperature < 32 :
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    elif temperature > 70:
        return 'Hot'
    else:
        return '-'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

air_data
```

```python
unhealthy_pm_10 = air_data[air_data['pm10.0_atm'] > 155]

unhealthy_events_pm10 = unhealthy_pm_10[
    ['date', 'sensor.name', 'pm10.0_atm']].sort_values('pm10.0_atm')

unhealthy_events_pm10
```
