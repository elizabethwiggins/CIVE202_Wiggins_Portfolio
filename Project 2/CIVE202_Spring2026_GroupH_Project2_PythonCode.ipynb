```python
# Defining fundamental constants used throughout the mix design

cubic_yard_ft3=27 #cubic feet in 1 cubic yard
unit_weight_water = 62.4 #lb/ft^3 (unit weight of water)
```

```python
# defining a function to calculate water weight
def calculate_water_weight_Q(cement_A, fly_ash_B, silica_C, other_scm_D, wc_ratio_E):
   
    #equation for water weight
    water_weight_Q = (cement_A+ fly_ash_B + silica_C + other_scm_D)*wc_ratio_E

    # this returns water weight
    return water_weight_Q

```

```python
# Defining a function to calculate the volume of cement (R)

def calculate_volume_R(cement_A, sg_cement_J):
    volume_cement_R = cement_A/(sg_cement_J * unit_weight_water)
    return volume_cement_R
    
```

```python
# Defining a function to calculate fly ash volume

def calculate_volume_S(fly_ash_B, sg_fly_ash_K):
    fly_ash_volume_S = fly_ash_B/(sg_fly_ash_K*unit_weight_water)
    return fly_ash_volume_S
```

```python
# Defining a function to calculate silica fume volume

def calculate_volume_T(silica_C, sg_silica_L):
    silica_volume_T = silica_C/(sg_silica_L*unit_weight_water)
    return silica_volume_T
```

```python
# Defining a function to calculate other SCM volume


def calculate_volume_U(other_scm_D, sg_other_scm_M):
    other_volume_U = other_scm_D/(sg_other_scm_M*unit_weight_water)
    return other_volume_U
```

```python
# Defining a function to calculate other air volume
#air content needs to be input as a percent
def air_volume_V(air_content_F):
    volume_V = (air_content_F/100) * cubic_yard_ft3
    return volume_V
    
```

```python
# Defining a function to calculate water volume (W) as a function of water weight (Q)
def calculate_water_volume_W(water_weight_Q, unit_weight_water):
    water_volume_W = water_weight_Q/unit_weight_water
    return water_volume_W
```

```python
# Defining a function to calculate total aggregate volume (X)
def calculate_aggregate_volume_X(volume_R, volume_S, volume_T, volume_U, volume_V, volume_W):
    volume_X = (cubic_yard_ft3 - volume_R- volume_S - volume_T - volume_U - volume_V - volume_W)
    return volume_X
    
```

```python
# Defining a function to calculate fine aggregate weight (Y)
def calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, aggregate_volume_X, unit_weight_water):
    fine_aggregate_Y= (unit_weight_water*(percent_fine_G/100)*sg_fine_N*aggregate_volume_X)
    return fine_aggregate_Y
```

```python
# Function for coarse aggregate weight (Z)

def calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, aggregate_volume_X, unit_weight_water):
    coarse_aggregate_Z= (unit_weight_water*(percent_coarse_H/100)*sg_coarse_O*aggregate_volume_X)
    return coarse_aggregate_Z
```

```python
# Function for other aggregate weight (AA)

def calculate_other_aggregate_AA(percent_other_I, sg_other_P, aggregate_volume_X, unit_weight_water):
    other_aggregate_AA= (unit_weight_water*(percent_other_I/100)*sg_other_P*aggregate_volume_X)
    return other_aggregate_AA
```

```python
# NDOT Concrete Mix Design – Full Integrated Run

# Project metadata
project_number = int(input("Enter project number: "))
class_of_concrete = input("Enter class of concrete (e.g., 47B, 47BR): ")

# Cementitious material inputs (lb per cubic yard)
cement_A = float(input("Enter cement weight A (lb per cubic yard): "))
fly_ash_B = float(input("Enter fly ash weight B (lb per cubic yard): "))
silica_fume_C = float(input("Enter silica fume weight C (lb per cubic yard): "))
other_scm_D = float(input("Enter other SCM weight D (lb per cubic yard): "))

water_cement_ratio_E = float(input("Enter target water-cement ratio E: "))
air_content_F = float(input("Enter target air content F (%): "))

# Aggregate proportions (%)
percent_fine_G = float(input("Enter percent fine aggregate G (%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H (%): "))
percent_other_I = float(input("Enter percent other aggregate I (%): "))

# Specific gravities
sg_cement_J = float(input("Enter specific gravity of cement J: "))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K: "))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L: "))
sg_other_scm_M = float(input("Enter specific gravity of other SCM M: "))

sg_fine_N = float(input("Enter specific gravity of fine aggregate N: "))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O: "))
sg_other_P = float(input("Enter specific gravity of other aggregate P: "))

# Calculations
Q = calculate_water_weight_Q(
    cement_A, fly_ash_B, silica_fume_C, other_scm_D, water_cement_ratio_E
)

R = calculate_volume_R(cement_A, sg_cement_J)
S = calculate_volume_S(fly_ash_B, sg_fly_ash_K)
T = calculate_volume_T(silica_fume_C, sg_silica_fume_L)
U = calculate_volume_U(other_scm_D, sg_other_scm_M)

V = air_volume_V(air_content_F)
W = calculate_water_volume_W(Q, unit_weight_water)
X = calculate_aggregate_volume_X(R, S, T, U, V, W)

Y = calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, X, unit_weight_water)
Z = calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, X, unit_weight_water)
AA = calculate_other_aggregate_AA(percent_other_I, sg_other_P, X, unit_weight_water)

# Output: Final Weight Chart
print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")
print("---------------------------------------------")
print(f"Project Number:       {project_number}")
print(f"Class of Concrete:    {class_of_concrete}")
print("---------------------------------------------")
print(f"Cement (A):            {cement_A:8.1f} lb")
print(f"Fly Ash (B):           {fly_ash_B:8.1f} lb")
print(f"Silica Fume (C):       {silica_fume_C:8.1f} lb")
print(f"Other SCM (D):         {other_scm_D:8.1f} lb")
print("---------------------------------------------")
print(f"Fine Aggregate (Y):    {Y:8.0f} lb")
print(f"Coarse Aggregate (Z):  {Z:8.0f} lb")
print(f"Other Aggregate (AA):  {AA:8.0f} lb")
print("---------------------------------------------")
print(f"Water (Q):             {Q:8.0f} lb")
print("---------------------------------------------")
print("End of Mix Design Summary")
```
